<h3><li>Lendo arquivo contendo os dados

```python
import pandas as pd

url ='https://raw.githubusercontent.com/AdrielGoulart/Breast-Cancer-Wisconsin-Diagnostic-Data-Set/master/data.csv'

df = pd.read_csv(url, delimiter=',',header = 0)
df.head()
```

<h2>1 - Limpeza de dados

### Seleção de Atributos Relevantes

<ul>
    
<li><u>diagnosis</u>: Diagnóstico da severidade do câncer(M=maligno, B=benigno)</li><br>
    
<li><u>radius_mean</u>: Média das distâncias do centro aos pontos no perímetro.</li><br>
    
<li><u>texture_mean</u>: Desvio padrão dos valores da escala de cinza.</li><br>
    
<li><u>perimeter_mean</u>: Tamanho médio do tumor central.</li><br>
    
<li><u>area_mean</u></li><br>
    
<li><u>smoothness_mean</u>: Média da variação local nos comprimentos do raio.</li><br>
    
<li><u>compactness_mean</u>: Valor do perímetro ^ 2 / área - 1,0.</li><br>
    
<li><u>concavity_mean</u>: Média de gravidade das porções côncavas do contorno.</li><br>
    
<li><u>concave points_mean</u>: Média para o número de porções côncavas do contorno.</li><br>
    
<li><u>symmetry_mean</u>: </li><br>
    
<li><u>fractal_dimension_mean</u>: Média para "coastline approximation" - 1.</li><br>
    
    
Esses atributos foram escolhidas devido serem as principais, ou seja as que apontam as dimensões e características dos tumores com maior relevância, 
pois possuem sua área a suavidade média, a distância entre o centro aos pontos do perímetro e também pelo dataset frizar que estes são os principais dados, os demais atributos não 
foram escolhidos pois muitos deles eram obtidos a partir de outros atributos ou simplesmente não possuiam uma descrição
legível.

<h4><li>Excluindo atributos e mostrando a nova tabela

```python
#Excluindo colunas(atributos)

df.drop('id', inplace=True, axis=1)

#df.drop('compactness_mean', inplace=True, axis=1)

#df.drop('concavity_mean', inplace=True, axis=1)

#df.drop('concave points_mean', inplace=True, axis=1)

df.drop('texture_se', inplace=True, axis=1)

df.drop('perimeter_se', inplace=True, axis=1)

df.drop('smoothness_se', inplace=True, axis=1)

df.drop('area_se', inplace=True, axis=1)

df.drop('compactness_se', inplace=True, axis=1)

df.drop('concavity_se', inplace=True, axis=1)

df.drop('concave points_se', inplace=True, axis=1)

df.drop('symmetry_se', inplace=True, axis=1)

df.drop('fractal_dimension_se', inplace=True, axis=1)

df.drop('radius_worst', inplace=True, axis=1)

#df.drop('fractal_dimension_mean', inplace=True, axis=1)

#df.drop('symmetry_mean', inplace=True, axis=1)

df.drop('radius_se', inplace=True, axis=1)

df.drop('texture_worst', inplace=True, axis=1)

df.drop('perimeter_worst', inplace=True, axis=1)

df.drop('area_worst', inplace=True, axis=1)

df.drop('smoothness_worst', inplace=True, axis=1)

df.drop('compactness_worst', inplace=True, axis=1)

df.drop('fractal_dimension_worst', inplace=True, axis=1)

df.drop('concavity_worst', inplace=True, axis=1)

df.drop('concave points_worst', inplace=True, axis=1)

df.drop('symmetry_worst', inplace=True, axis=1)

df.drop('Unnamed: 32', inplace=True, axis=1)


## Mostrando a nova tabela gerada
df.head()
```

<h2>2 - Transformação de Dados

<h3><li> Normalização de Dados Numéricos
    <h4> Como os dados dos atributos estão variando muito decidiu-se normalizar o conjunjo de dados

```python
df.head()
```

```python
#Funções
import numpy as np
def normalizacao_linear(x): 
    return (x - np.min(x)) / (np.max(x) - np.min(x));
```

```python
# Normalizando cada uma das colunas
#normalizacao_df = df.copy() 

df['radius_mean'] = normalizacao_linear( df['radius_mean'] )

df['texture_mean'] = normalizacao_linear( df['texture_mean'] )

df['perimeter_mean'] = normalizacao_linear( df['perimeter_mean'] )

df['area_mean'] = normalizacao_linear( df['area_mean'] )

df['smoothness_mean'] = normalizacao_linear( df['smoothness_mean'] )

df.head()
```
